```python
import pandas as pd

pd.set_option("future.no_silent_downcasting", True)  # noqa: FBT003
```

```python
train_row_df = pd.read_csv("train.csv")
test_row_df = pd.read_csv("test.csv")
# pd.plotting.scatter_matrix(train_row_df)  # noqa: ERA001
```

```python
train_df = train_row_df.drop(
    columns=["PassengerId", "Name", "Ticket", "Cabin", "Embarked"]
)
test_df = test_row_df.drop(
    columns=["PassengerId", "Name", "Ticket", "Cabin", "Embarked"]
)
```

```python
train_df["Sex"] = train_df["Sex"].replace(["male", "female"], [0, 1]).astype("int")
test_df["Sex"] = test_df["Sex"].replace(["male", "female"], [0, 1]).astype("int")
```

```python
train_df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train = train_df.iloc[:, 1:]
Y_train = train_df.iloc[:, 0]
```

```python
test_df.isna().sum()
```

```python
age_median = X_train["Age"].median()
fare_median = X_train["Fare"].median()
X_train["Age"] = X_train["Age"].fillna(age_median)

test_df["Age"] = test_df["Age"].fillna(age_median)
test_df["Fare"] = test_df["Fare"].fillna(fare_median)
```

```python
clf = RandomForestClassifier(random_state=0)
clf.fit(X_train, Y_train)
```

```python
y_pre = clf.predict(test_df)
```

```python
y_pre
```

```python
gender_submisstion = pd.read_csv("titanic/gender_submission.csv")
```

```python
sub = gender_submisstion
sub["Survived"] = list(map(int, y_pre))
sub.to_csv("submission.csv", index=False)
```

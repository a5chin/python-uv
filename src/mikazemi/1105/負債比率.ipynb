```python
df_fusai_raw = pd.read_csv("負債比率2016-2021(1).csv", encoding="cp932")
df_fusai_raw2 = pd.read_csv("fusai2022.csv", encoding="cp932")
df_fusai_raw2
```

```python
df_fusai_drop_renketsu = df_fusai_raw.drop(df_fusai_raw.columns[74:], axis=1)
df_fusai_raw2.columns
```

```python
df_fusai_drop_row = df_fusai_drop_renketsu.drop(
    df_fusai_drop_renketsu.index[[0, 1, 2, 3, 4, 5, 6, 8, 9]]
)
df_fusai_drop_row2 = df_fusai_raw2.drop(
    df_fusai_drop_renketsu.index[[0, 1, 2, 3, 4, 5, 6, 8, 9]]
)

df_fusai_merge = df_fusai_drop_row.merge(df_fusai_drop_row2, on="Unnamed: 1")
df_fusai_merge
```

```python
df_fusai_drop_cumpanyname = df_fusai_merge.drop(
    columns=df_fusai_merge.columns[0]
).reset_index(drop=True)


df_fusai_drop_cumpanyname
```

```python
df_fusai_drop_cumpanyname.iloc[0, 0] = "code"
df_fusai_drop_cumpanyname
```

```python
colum_list = df_fusai_drop_cumpanyname.iloc[0, :]
colum_list
```

```python
df_change_colums = df_fusai_drop_cumpanyname.drop(
    index=df_fusai_drop_cumpanyname.index[0]
).reset_index(drop=True)
df_change_colums.columns = colum_list
df_change_colums
```

```python
df_fusai_drop_code_t = df_change_colums
df_fusai_drop_code_t["code"] = df_fusai_drop_code_t["code"].str.replace("T", "")
df_fusai_drop_code_t
```

```python
df_fusai_melt = df_fusai_drop_code_t.melt(
    id_vars="code", var_name="year/month", value_name="fusaiRate"
)
df_fusai_melt
```

```python
# df_fusai_melt["year"] = pd.to_datetime(
#     df_fusai_melt["year/month"], format="%Y/%m"  # noqa: ERA001
# ).dt.year
# df_fusai_melt
```

```python
df_fusai_melt_drop = df_fusai_melt[
    (~df_fusai_melt["year/month"].isin(["2016/01", "2016/02", "2016/03"]))
    & (df_fusai_melt["year/month"] < "2022/04")
]
df_fusai_melt_drop
```

```python
# 年度列を追加
def get_fiscal_year(ym):  # noqa: ANN201, D103, ANN001
    year, month = map(int, ym.split("/"))
    return year if month >= 4 else year - 1  # noqa: PLR2004


# コピーして新しい名前に
df_fusai_melt_drop_safe = df_fusai_melt_drop.copy()
df_fusai_melt_drop_safe["fiscal_year"] = df_fusai_melt_drop_safe["year/month"].apply(
    get_fiscal_year
)
df_fusai_melt_drop_safe["fusaiRate"] = pd.to_numeric(
    df_fusai_melt_drop_safe["fusaiRate"], errors="coerce"
)

# 年度ごとに平均を集計
df_fiscal_group = (
    df_fusai_melt_drop_safe.groupby(["code", "fiscal_year"])["fusaiRate"]
    .mean()
    .reset_index()
)
df_fiscal_group
```

```python
df_fusai_dropna = df_fiscal_group.dropna()
```

```python
df_fusai_dropna.to_csv("負債比率_年度別")
```

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_excel("data_前半.xlsx", sheet_name="Sheet1")
df2 = pd.read_excel("data_後半.xlsx", sheet_name="Sheet1")
```

```python
df1_1 = df1.iloc[:, 0:5]
df2_2 = df2.iloc[:, 0:5]
```

```python
df3 = pd.concat((df1_1, df2_2)).dropna()
df3_1 = df3.replace("-", np.nan).dropna()
df3_1
```

```python
df3_1.astype(
    {
        "資産合計": "int64",
        "自己資本／親会社の所有者に帰属する資本": "int64",  # noqa: RUF001
        "【株主資本合計／親会社の所有者に帰属する資本】親会社株主に帰属する当期純利益（連結）／当期純利益（単独）": "int64",  # noqa: E501, RUF001
    }
)
```

```python
df3_1["ROA"] = df3_1.iloc[:, 4] / df3_1.iloc[:, 2] * 100
df3_1["ROE"] = df3_1.iloc[:, 4] / df3_1.iloc[:, 3] * 100
```

```python
df3_1.columns = ["会社名", "東証コード", "資産合計", "自己資本", "純利益", "ROA", "ROE"]
```

```python
df3_1["東証コード"] = df3_1["東証コード"].str.strip("T")
df3_1["東証コード"] = df3_1["東証コード"].str.strip("A")
```

```python
df3_1
```

```python
df3_1.columns = ["会社名", "証券コード", "資産合計", "自己資本", "純利益", "ROA", "ROE"]
df3_2 = df3_1.iloc[:, 1:]
```

```python
df3_2.to_excel("df3_2.xlsx")
```

```python
df_teuchi = pd.read_csv("mika_teuti_data.csv")
```

```python
df3_3 = df3_2.astype({"証券コード": "float64"})
```

```python
df3_3.dtypes
```

```python
df_kansei_data = df_teuchi.merge(df3_3, on="証券コード")
df_perfect_data = df_kansei_data.iloc[:, 1:]
df_perfect_data
```

```python
df_perfect_data["従業員の能力・評価結果の公開"] = df_perfect_data[
    "従業員の能力・評価結果の公開"
].replace(3, np.nan)
```

```python
df_perfect_data["有休取得率/前期"] = df_perfect_data["有休取得率/前期"].where(
    df_perfect_data["有休取得率/前期"] < 100  # noqa: PLR2004
)
```

```python
df_perfect_data = df_perfect_data.dropna()
```

```python
df_perfect_data.iloc[:, 1:].describe().to_excel("各変数の基本統計量.xlsx")
```

```python
df_perfect_data.iloc[:, [1, 2, 3, 4, 5, 6, 7, 8, 9, 13, 14]].corr().to_excel(
    "各変数の相関.xlsx"
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.heatmap(
    df_perfect_data.iloc[:, [1, 2, 3, 4, 5, 6, 7, 8, 9, 13, 14]].corr(),
    vmax=1,
    vmin=-1,
    center=0,
)
plt.savefig("各変数の相関.png", bbox_inches="tight")
plt.close("all")
```
